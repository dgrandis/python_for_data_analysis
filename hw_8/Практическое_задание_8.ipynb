# Импортирование библиотек

```python
import requests
import pandas as pd
import psycopg2
```

# **Работа с API**

## 1. Изучение документации [API OF ICE AND FIRE](https://github.com/joakimskoog/AnApiOfIceAndFire/wiki)

## 2. Получить информацию обо всех книгах и сохранить в pd.DataFrame - **1 балл**

```python
# Базовый url
api_base_url = 'https://anapioficeandfire.com/'
```

```python
# В начале посмотрим какой статус ответа
book_list = requests.get(
    f'{api_base_url}/api/books')

book_list
```

```python
# Сохранение данных в формате json
book_list = requests.get(
    f'{api_base_url}/api/books').json()
```

```python
# Сохранение в DataFrame
pd_books = pd.DataFrame(book_list)

# Первые 5 строк таблицы
pd_books.head()
```

## 3. Получить информацию обо всех домах Вестероса (Houses) и сохранить в pd.DataFrame - **1 балл**

```python
# Сохранение всех домов в формате json
houses_list = requests.get(
    f'{api_base_url}/api/houses').json()

# Пример первого дома
houses_list[0]
```

```python
# Сохранение в DataFrame
pd_houses = pd.DataFrame(houses_list)
```

```python
# Первые 5 строк таблицы
pd_houses.head()
```

## 4. Получить информацию обо всех домах Вестероса, у которых есть девиз, и сохранить в pd.DataFrame - **2 балла**

```python
# Сохранение всех домов в формате json, использовав флаг hasTitles равный True
houses_hastitles_list = requests.get(
    f'{api_base_url}/api/houses?hasTitles=True').json()

# Пример первого дома с девизом
houses_hastitles_list[0]
```

```python
# Сохранение в DataFrame
pd_houses_hastitles = pd.DataFrame(houses_hastitles_list)
```

```python
# Первые 5 строк таблицы
pd_houses_hastitles.head()
```
