## 4. Выведите самое популярное мужское и самое популярное женское имя на корабле - **1 балл**


```python
# Список мужских имен
man_name = df[df['Sex'] == 'male']['Name']

# Список женских имен
woman_name = df[df['Sex'] == 'female']['Name']
```

```python
# Пример списка женских имен
woman_name.value_counts()
```

Как видно у замужних женщин чаще всего после "Mrs." стоит имя мужа, а имя и фамилия женщины располагаются в скобках.

Поэтому сначала парсить имена замужних женщин из скобок, а потом все оставшиеся.

```python
def parse_name(name):
    first = name.str.extract(r"Mrs\.\s+[^(]*\((\w+)", expand=False)
    first.loc[first.isna()] = name.str.extract(r"\.\s+(\w+)", expand=False)
    return first
```

```python
popular_male_name  = parse_name(man_name).value_counts()
popular_male_name
```

```python
popular_female_name  = parse_name(woman_name).value_counts()
popular_female_name
```

```python
print('Самое популярное мужское имя:', popular_male_name.index[0])
print('Самое популярное женское имя:', popular_female_name.index[0])
```

```python
# В фильтрацию добавим класс пассажиров на каждой иттерации  и воспользуемся функцией объявленной в прошлом задании
for pclass in df['Pclass'].unique():

    # Список имен по в зависимости от класса и пола
    man_name = df[(df['Pclass'] == pclass) & (df['Sex'] == 'male')]['Name']
    woman_name = df[(df['Pclass'] == pclass) & (df['Sex'] == 'female')]['Name']

    # Самые популярные имена по количеству
    popular_male_name  = parse_name(man_name).value_counts()
    popular_female_name  = parse_name(woman_name).value_counts()

    # Выводим все имена (ключи) с максимальным значением, так как может быть одинаковое количество имен в одном классе
    max_value_male = popular_male_name.max()
    max_value_female = popular_female_name.max()

    # Список таких имен
    max_names_male = popular_male_name[popular_male_name == max_value_male].index.tolist()
    max_names_female = popular_female_name[popular_female_name == max_value_female].index.tolist()

    print(f'{pclass} класс:')
    print(f'Самое популярное мужское имя в классе : {max_names_male}, количество: {max_value_male}')
    print(f'Самое популярное женское имя в классе : {max_names_female}, количество  {max_value_female}')
    print()
```

```python
# Создаем pandas.Series
s = pd.Series({'Elizabeth': 5, 'Margaret': 5, 'Mary': 4, 'Alice': 4, 'Anna': 3})

# Находим максимальное значение в серии
max_value = s.max()

# Выводим все имена (ключи) с максимальным значением
max_names = s[s == max_value].index.tolist()
print(max_names)
```

## 8.Выведите количества n-местных кабин (в которых было 2, 3, 4, ... человека) -**1 балл**

```python
# Исходить будем из столбца SibSp - Братьев и сестер на борту
sibSp_unique = df.SibSp.unique()
sibSp_unique
```

Соответственно значение:
 - 0 - кто путешествовал один
 - 1 - вдвоем
 - 2 - втроем
 - и т.д.

```python
# Список пассажир кто путешествовал не один, то есть были в n- местных каютах
more_than_one = df[df['SibSp'] != 0]
more_than_one
```

```python
# Значит можно считать что те кто были не одни, у тех были n-местные номера
more_than_one_counts = more_than_one['SibSp'].value_counts()

# И отсортируем по индексам
more_than_one_counts = more_than_one_counts.sort_index()
```

```python
for index, item in more_than_one_counts.items():
  print(f'Число {index+1}-местных кабин: {item}')
```
